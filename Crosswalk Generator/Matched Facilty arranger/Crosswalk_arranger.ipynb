```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jul  7 13:06:19 2017

@author: tghosh

"""

import matplotlib.pyplot as plt
import numpy as np
import openpyxl
from openpyxl.styles import Color, PatternFill, Font, Border
from openpyxl.styles import colors
import os
import statistics


print('Opening workbook...')

#Its easy if you run the code with each block by block again. Fine otherwise too. 


#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################
#Reading the files and the sheets
wb = openpyxl.load_workbook('egridtoeis.xlsx')
sheet1 = wb.get_sheet_by_name('egridtoeis')

#Reading a blank file for writing purposes. SHould have blank Sheet1 present
wb2 = openpyxl.load_workbook('chk.xlsx')
sheet = wb2.get_sheet_by_name('Sheet1')


FRS_Code = sheet1['A2':'A2668']
Name = sheet1['D2':'D2668']
Egrid_ID = sheet1['F2':'F2668']
Eis_ID = sheet1['C2':'C2668']


# This part of the code essentially looks for repeating EIS iDS for a same EGRID ID and then writes them in columns rather than
# in different rows. Check output file. 

s = 0;
br =[];
p = 1;
i = 0;
while i < 2667:
    s = Name[i][0].value;
    ar =[];
    for j in range(i,2667):
        if s == Name[j][0].value:
            ar.append(Eis_ID[j][0].value)
        else:
            break;
    sheet.cell(row = p,column = 1).value = s;
    sheet.cell(row = p,column = 2).value = FRS_Code[i][0].value;
    sheet.cell(row = p,column = 3).value = Egrid_ID[i][0].value;
    for k in range(0,len(ar)):
       sheet.cell(row = p,column = 4+k).value = ar[k]; 
    i = j;
    p=p+1;
    if i == 2666:
        break;
    
wb2.save('chk1.xlsx')
#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################
#Same operation for the other databases
wb = openpyxl.load_workbook('egridtonpdes.xlsx')
sheet1 = wb.get_sheet_by_name('egridtonpdes')


wb2 = openpyxl.load_workbook('chk.xlsx')
sheet = wb2.get_sheet_by_name('Sheet1')


FRS_Code = sheet1['A2':'A2678']

Name = sheet1['D2':'D2678']

Egrid_ID = sheet1['F2':'F2678']

npdes_ID = sheet1['C2':'C2678']

s = 0;

br =[];
p = 1;
i = 0;
while i < 2677:
    s = Name[i][0].value;
    ar =[];
    for j in range(i,2677):
        if s == Name[j][0].value:
            ar.append(npdes_ID[j][0].value)
        else:
            break;
    sheet.cell(row = p,column = 1).value = s;
    sheet.cell(row = p,column = 2).value = FRS_Code[i][0].value;
    sheet.cell(row = p,column = 3).value = Egrid_ID[i][0].value;
    for k in range(0,len(ar)):
       sheet.cell(row = p,column = 4+k).value = ar[k]; 
    i = j;
    p=p+1;
    if i == 2676:
        break;
    
wb2.save('chk2.xlsx')  

#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################


wb = openpyxl.load_workbook('egridtotri.xlsx')
sheet1 = wb.get_sheet_by_name('egridtotri')


wb2 = openpyxl.load_workbook('chk.xlsx')
sheet = wb2.get_sheet_by_name('Sheet1')


FRS_Code = sheet1['A2':'A1224']

Name = sheet1['D2':'D1224']

Egrid_ID = sheet1['F2':'F1224']

tri_ID = sheet1['C2':'C1224']

s = 0;

br =[];
p = 1;
i = 0;
while i < 1223:
    s = Name[i][0].value;
    ar =[];
    for j in range(i,1223):
        if s == Name[j][0].value:
            ar.append(tri_ID[j][0].value)
        else:
            break;
    sheet.cell(row = p,column = 1).value = s;
    sheet.cell(row = p,column = 2).value = FRS_Code[i][0].value;
    sheet.cell(row = p,column = 3).value = Egrid_ID[i][0].value;
    for k in range(0,len(ar)):
       sheet.cell(row = p,column = 4+k).value = ar[k]; 
    i = j;
    p=p+1;
    if i == 1222:
        break;
    
wb2.save('chk3.xlsx')   

#########################################****************************************************************##############################################################
#########################################****************************************************************##############################################################
```
